## Distributions

In statistics and probability theory, a distribution refers to the **set of all possible values and their corresponding probabilities or frequencies of occurrence** for a given variable or set of variables. In other words, it is a way of *describing the spread of data in a dataset*.

**Probability Distribution**: A probability distribution is a *mathematical function* that describes the likelihood of obtaining different possible outcomes from a random variable.


_For example, if we roll a fair six-sided die, the random variable is the number that comes up, and the probability distribution describes the probability of rolling each number, which is 1/6 for each number._


Probability distributions can be described using different mathematical functions, including **probability density functions (PDFs) for continuous distributions and probability mass functions (PMFs) for discrete distributions**.


Some common examples of probability distributions include the *normal distribution, the binomial distribution, the Poisson distribution, the exponential distribution, and the uniform distribution*. These distributions are used in various applications, including **statistical inference, hypothesis testing, and machine learning**.


**Different distributions:**

- **Normal Distribution**: Also known as the Gaussian distribution, this is a **bell-shaped distribution that is symmetric around the mean**. It is commonly used in *statistical analysis and modeling* due to its simple and predictable nature. It has a *single peak at the mean and the curve gradually decreases as you move away from the mean*.


- **Binomial Distribution**: This distribution is used when there are **only two possible outcomes for a given event**, such as success or failure, or heads or tails. It is commonly used in *binary classification problems* in Machine Learning. The graph of a binomial distribution is **usually a symmetrical bell-shaped curve for large sample sizes**. It has a *single peak and is centered around the mean*. For *small sample sizes, it may be skewed*.

- **Poisson Distribution**: This distribution is used to **model the probability of rare events occurring**. It is commonly used in *count-based data analysis*, such as the number of customers visiting a store in a given day. The graph of a Poisson distribution is a **skewed right curve with a long tail**. It is *unimodal and its shape changes with the mean*.


- **Exponential Distribution**: This distribution is used to **model the time between events that occur at a constant rate**. It is commonly used in *survival analysis and reliability modeling*. The graph of an exponential distribution is a **skewed right curve that starts at the origin and gradually increases**. It has a *single peak and is used to model waiting times*.


- **Uniform Distribution**: This distribution is used to **model data where each value has an equal probability of occurring**. It is *commonly used in random number generation and simulation*. The graph of a uniform distribution is a **rectangular-shaped curve with equal height across the range of values**. It has **no peaks or valleys** and is used to *model random variables that are equally likely to take on any value within a given range*.


### Different distributions are helpful in Machine Learning in a variety of ways. Some of the most common applications include:


- **Data Preprocessing**: Understanding the distribution of data in a dataset is crucial for effective data preprocessing. For example, if the data follows a normal distribution, we can use techniques such as normalization to make the data more suitable for modeling.


- **Feature Engineering**: Understanding the distribution of features can help in feature engineering, where we **transform and combine features to improve model performance**. For example, we *can transform skewed features using logarithmic or exponential functions to make them more suitable for modeling*.


- **Model Selection**: Different Machine Learning models assume different distributions of data. For example, **linear regression assumes that the data follows a normal distribution**, while decision trees do not make any assumptions about the distribution of data. *Understanding the distribution of data can help in selecting the appropriate model for a given problem*.


- **Model Evaluation**: *Different evaluation metrics assume different distributions of data.* For example, the **mean squared error (MSE) assumes that the data follows a normal distribution**, while the **area under the receiver operating characteristic (ROC) curve does not make any assumptions about the distribution of data**. Understanding the distribution of data can help in *selecting appropriate evaluation metrics for a given problem*.

Overall, understanding different types of distributions and their applications in Machine Learning can help in developing more accurate and effective models.

## Seaborn for Distribution plots

Seaborn provides several methods to plot different types of distribution plots, which are commonly used in exploratory data analysis (EDA) and visualization of data. 

**Some of the popular Seaborn methods to plot distribution plots are:**


- **distplot()**: This method *plots a histogram and an estimate of the probability density function (PDF)* of a **continuous variable**. It also provides options to customize the plot, such as setting the number of bins, choosing the color and style of the plot, and *adding a kernel density estimate (KDE)*.


- **kdeplot()**: This method *plots a **kernel density estimate (KDE)** of a continuous variable*, which is a *non-parametric way of estimating the probability density function (PDF)*.


- **rugplot()**: This method plots a *rug plot of a continuous variable*, which **shows the individual data points as tick marks along the x-axis**. It can be used to *visualize the distribution of data and to compare it with a KDE or a histogram*.

- **violinplot()**: This method plots a violin plot of a **continuous variable or a categorical variable**, which shows the **shape of the distribution as a kernel density estimate (KDE) and a box plot**. It can be used to *visualize the relationship between two variables and to detect any patterns or trends in the data*.

### Histogram and distplots

**Histograms are commonly used in exploratory data analysis (EDA) and machine learning for the following purposes**:

- **Visualizing the distribution of data**: Histograms can provide a **visual summary of the distribution of data**, including the *shape, center, and spread of the data*. They can help to **identify any skewness, outliers, or other anomalies in the data**, and to compare the distribution of data across different groups or categories.


- **Determining the optimal number of bins**: Histograms can be used to **determine the optimal number of bins**, which is the number of intervals into which the data is divided. Choosing the optimal number of bins is important to ensure that the histogram accurately reflects the distribution of the data.


- **Detecting patterns or trends in the data**: Histograms can be used to detect any patterns or trends in the data, such as a *shift in the center or a change in the spread of the data over time*. They can help to **identify any relationships or correlations between variables**, and to **test hypotheses** about the causal mechanisms underlying these relationships.


- **Identifying outliers or anomalies in the data**: Histograms can be used to identify any outliers or anomalies in the data, which are data points that are significantly different from the rest of the data. They can help to *identify any errors or omissions in the data, and to determine whether to include or exclude these data points in the analysis*.

```python
sns.distplot(iris.sepal_length)
plt.show()
```

```python
sns.histplot(iris.sepal_length)
plt.show()
```

```python
sns.distplot(iris.petal_length);
```

```python
sns.histplot(iris.petal_length);
```

```python
sns.histplot(x='petal_length', data=iris, hue='species')
plt.show()
```

### KDE (Kernel Density Estimator) plots

**Smoothed out version of histogram**

```python
sns.kdeplot(iris.sepal_length)
plt.show()
```

```python
sns.kdeplot(iris.sepal_length, fill=True)
plt.show()
```

```python
sns.kdeplot(iris.petal_length, fill=True);
```

```python
### KDE plot based on species

sns.kdeplot(x='petal_length', data=iris, fill=True, hue='species')
plt.show()
```

**Observation:**

- One can Identify Iris Setosa just by looking at petal length. If it is less than 2cm, it is most likely to be Setosa.
- Petal length for versicolor and virginica are almost identical, with virginica being about 1cm long on average

### Plotting 2D KTE plot

## Box Plots

**Boxplots are commonly used in exploratory data analysis (EDA) and machine learning for the following purposes:**

- **Visualizing the distribution of data**: Boxplots can provide a *visual summary of the distribution of data*, including the **median, quartiles, and outliers**. They can help to *identify any skewness, outliers, or other anomalies* in the data, and to compare the distribution of data across different groups or categories.


- **Comparing the distribution of data across different groups or categories**: Boxplots can be used to compare the distribution of data across different groups or categories, such as comparing the distribution of test scores between male and female students, or comparing the distribution of income across different states or regions. *They can help to identify any differences or similarities in the distribution of data, and to test hypotheses about the relationship between variables*.


- **Detecting patterns or trends in the data**: Boxplots can be used to detect any patterns or trends in the data, such as a **shift in the median or a change in the spread of the data over time**. They can help to *identify any relationships or correlations between variables, and to test hypotheses about the causal mechanisms underlying these relationships*.


- **Identifying outliers or anomalies in the data**: Boxplots can be used to identify any outliers or anomalies in the data, which are data points that are significantly different from the rest of the data. They can help to *identify any errors or omissions in the data, and to determine whether to include or exclude these data points in the analysis*.


In machine learning, boxplots can be used to **pre-process and visualize data** before applying various algorithms to it. For example, boxplots can help to **identify any missing values or outliers in the data**, which can be addressed by either **imputing missing values or removing the outliers**. They can also be used to *compare the distribution of data across different classes or labels*, which can help to **determine the best algorithm to use for the problem at hand**.

**Interpretation of Box Plots**:

*Here are some key points to consider when interpreting a box plot*:

- **Median**: The horizontal line within the box represents the median value of the data.


- **Box**: The box itself represents the **interquartile range (IQR)**, which is the range of values that contains 50% of the data. The lower edge of the box represents the **first quartile (Q1)**, while the upper edge represents the **third quartile (Q3)**.


- **Whiskers**: The whiskers extend from the box and indicate the range of data within **1.5 times the IQR**. *Any data points outside of this range are considered outliers and are typically shown as individual points or dots*.


### Violin Plots

**Violin plots are commonly used in exploratory data analysis (EDA) and machine learning for the following purposes**:


- **Visualizing the distribution of data**: Violin plots provide a visual summary of the distribution of data, including the **shape, center, and spread of the data, similar to boxplots**. However, violin plots **also show the kernel density estimate of the data**, providing a more detailed view of the distribution of the data.


- **Comparing the distribution of data across different groups or categories**: Violin plots can be used to compare the distribution of data across different groups or categories, similar to boxplots. However, *violin plots provide a more detailed view of the distribution of the data, making it easier to detect any differences or similarities in the shape or spread of the data* across different groups or categories.


- **Identifying outliers or anomalies in the data**: Violin plots can be used to identify any outliers or anomalies in the data, similar to boxplots. However, violin plots **provide a more detailed view of the distribution of the data, making it easier to detect any anomalies** that may be hidden in the data.


- **Assessing the density of data**: Violin plots can be used to **assess the density of data, providing information on the frequency and intensity of data points across the range of values**. This can be particularly *useful for identifying any patterns or trends in the data that may be difficult to detect* using other visualization techniques.

*Interpreting a violin plot involves examining the **shape, spread, and density** of the data*. Here are some key points to consider when interpreting a violin plot:

- **Shape**: The shape of the violin plot indicates the distribution of the data. **A symmetrical shape indicates that the data is evenly distributed around the center**, while an **asymmetrical shape indicates that the data is skewed towards one side**. The shape can *also reveal any multimodality in the data, indicating the presence of multiple subpopulations*.


- **Width**: The **width of the violin plot indicates the density of the data at different points**. A *wider section indicates a higher density of data points*, while a *narrower section indicates a lower density of data points*.


- **Median**: The horizontal line within the violin plot indicates the median value of the data.


- **Interquartile range (IQR)**: The height of the violin plot indicates the range of values that contain 50% of the data (i.e., the IQR). The thicker the plot, the larger the IQR.


- **Outliers**: Any points outside of the violin plot, shown as points or dots, indicate outliers in the data.


- **Comparison**: When comparing multiple violin plots, differences in shape, spread, and density can be used to identify differences or similarities in the data across different groups or categories.


Overall, violin plots provide a more detailed view of the distribution of data compared to other visualization techniques such as histograms or boxplots, making it **easier to identify patterns or trends in the data**. They are particularly useful for comparing the distribution of data across different groups or categories and for detecting any anomalies or outliers in the data.

## Seasonal Decomposition [For time series data]

Seasonal decompose is a method used in time series analysis to **separate a time series into its underlying components of trend, seasonality, and residual**. It is used to *identify patterns in the data and to understand the relationship between the different components of the time series*.

The seasonal decompose method **uses mathematical techniques to break down a time series into its three main components: trend, seasonal variation, and residual variation**.

```python
### Loading a time series data

df = sns.load_dataset('flights')
df
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(df[['passengers']], period=12) # period 12 for getting monthly results
result.plot();
```

**Clearly these observations can be seen**:

- Upward trend in the data
- High Seasonality
- Residual do not looks random and look like mirrored image

## ACF and PACF charts [Used for Time Series]

These plots are *commonly used in time series analysis* to identify the **order of autoregressive (AR) and moving average (MA) models**. The order of the *AR model can be identified by looking at the last significant lag in the PACF plot*, while the *order of the MA model can be identified by looking at the last significant lag in the ACF plot*. The combined order of the AR and MA models can be identified by looking at both plots together.


**ACF (Autocorrelation Factor)** plot shows the **correlation of a time series with its own lagged values**. It calculates the *correlation between a time series and its lagged values at different lags*. The ACF plot can help to ***identify the presence of seasonality, trend and the lag at which the correlation is highest***. 


*If the plot shows a significant correlation for a specific lag value, it suggests that the value at that lag has an effect on the current value*.


**PACF (Partial Autocorrelation Factor)** plot, on the other hand, shows the **correlation between a time series and its lagged values after removing the effect of the intervening lags**. It helps to *identify the lag value where the correlation is significantly different from zero, and the number of lags that should be included in a time series model*. 

**A significant correlation at lag k indicates that the kth lag should be included in the model.**

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, ax = plt.subplots(2, figsize=(12, 12))
plot_acf(df[['passengers']], lags=30, ax=ax[0])
plot_pacf(df[['passengers']], lags=30, ax=ax[1]);
```

**Observations from ACF plot**:

- All lags are significant
- As the autocorrelation constant of lags kept on decreasing, a trend is there in the data
- Seasonality in data is also seen, as the acf pattern repeats itself every 12 lags (shows seasonality of 12)
- Data is non-stationary [When value drops to 0 after 1-2 lags, data is stationary]
- Data is not random [When all lags have correlation constant near 0]

**Observations from PACF plot**:

- We will get the idea on most significant lags here. **A significant bar in the PACF plot indicates that the lagged values have a direct impact on the current value of the time series.**
- All lags are not significant
- Lag 0 has perfect autocorrelation
- Lag 1 has highest correlation 
- Other lags like lag 2, 13, 22 are significant


**If a bar is significant in the ACF plot but not in the PACF plot, it suggests that the time series may be following a moving average model. If a bar is significant in the PACF plot but not in the ACF plot, it suggests that the time series may be following an autoregressive model.**


**If there are significant bars in both the ACF and PACF plots, it suggests that the time series may be following a mixed autoregressive-moving average (ARMA) model**. The order of the ARMA model can be determined by the number of significant bars in the plots and their positions.
