### Getting the ranking performance of the countries

```python
rankings = pd.DataFrame()

# The below code loop over all countries, counting the value frequency stored as a dataframe, then sorted and appended into 
# rankings dataframe.

for country in ['ARG', 'BRA', 'ESP', 'FRA', 'GER', 'ITA']:
    df_count_sorted = pd.DataFrame(df[str(country)].value_counts()).sort_index()
    rankings[str(country)] = df_count_sorted

# Filling the NaN values and converting the dataframe to integer type.
rankings.fillna(0, inplace=True)
rankings = rankings[rankings.columns].astype('int')
rankings
```

### Observations:

- Brazil has been at top for most of the times
- Also, Brazil is the best ranked team (mostly seen at top 6, rarely less rated)
- Spain performance is consistent after Brazil followed by Argentina
- Spain never had ranking beyond 16, shows its top consistency over years
- Among Germany, France and Italy, best performance is showed by Germany followed by France and then Italy

### Plotting the rank frequency histogram for each countries

```python
for i in [0, 2, 4]:
    
    # This code makes 2 subplots in a row, given both x and y grids will be of same shape with a given figure size.
    fig, ax = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(12, 4))
    
    # Histogram plot (Using .plot() function with kind='hist' for this)
    df[df.columns[i]].plot(kind='hist', ax=ax[0])
    df[df.columns[i+1]].plot(kind='hist', ax=ax[1])
    
    # Setting the x-label and y-label for 1st image
    ax[0].set_xlabel('Ranking', weight='bold')
    ax[0].set_ylabel('Frequency', weight='bold')
    
    # Setting the x-label and y-label for 2nd image
    ax[1].set_ylabel('Frequency', weight='bold')
    ax[1].set_xlabel('Ranking', weight='bold')
    
    # Setting title for both images of the row
    ax[0].set_title(f'Histogram for {df.columns[i]} ranking', weight='bold', size=15)
    ax[1].set_title(f'Histogram for {df.columns[i+1]} ranking', weight='bold', size=15)
    
    plt.show()
```

# Data Cleaning

### Checking if we have data for all the time periods [including every months of an year]

```python
# Method-1
# Resetting the index and counting, grouped by year

dfx = df.reset_index()
dfx.groupby([dfx.Date.dt.year]).Date.count()
```

```python
# Method-2
# value_count function on index 

df.index.year.value_counts()
```

### =============================================

**Following observations can be noted:**
- Starting and ending years have least observations (Okay, as data started and ended at particular intervals)
- 2 month data missing from 1994 to 1998
- 1 month data missing for 2002, 2006 and 2010

**Total 13 time-blocks missing**

```python
df[df.index.year == 1993]
```

```python
### Concept note for next code-block
## Comparing 2 lists and returning the non-matching element.


a = [1, 2, 5]
b = [1, 2]

# set.union(set(a), set(b)), set.intersection(set(a), set(b))
set(a) - set(b)
```

```python
### Checking for months having 2 missing values

months = list(range(1, 13))

for year in range(1994, 1999):
    print(year, df[df.index.year == year].index.month.values)
    print(year, sorted(list(set(months) - set(df[df.index.year == year].index.month.values))))
    print('---------')
```

```python
### Checking for months having 1 missing values

for year in [2002, 2006, 2010]:
    print(year, df[df.index.year == year].index.month.values)
    print(year, sorted(list(set(months) - set(df[df.index.year == year].index.month.values))))
    print('-----------')
```

```python
### There are two 3's in 2010 above, checking both months here

df[(df.index.year == 2010) & (df.index.month == 3)]
```

```python
### Checking missing months for all years

for year in df.index.year.unique()[1:-1]:
    if len(list(set(df[df.index.year == year].index.month.values))) != 12:
        print(year, sorted(list(set(months) - set(df[df.index.year == year].index.month.values))))
```

**So, overall 14 time-block values are missing**

But since there is a particular time these rankings are released, and not necessarily is every month, these missing time-blocks can be neglected.

```python
### Checking repeated months

for year in df.index.year.unique():
    if len(set(df[df.index.year == year].index.month.values)) != len(df[df.index.year == year].index.month.values):
        print(year)
```

So, only for 2010 the month is repeated

### Checking for duplicates

```python
df.reset_index().duplicated().sum()
```

**So, no duplicates**

### Data is all good to explore further

## Resampling data and getting all months outputs

Resampling by months and summing up values

```python
dfxx = df.resample('m').sum()
dfxx
```

```python
df.shape
```

**We see 13 values increases from shape of df in dfxx. This is because all months are now included after resampling, adding 14 missing months and removing 1 repeated month**

Checking for missing values

```python
dfxx.isna().sum()
```

Looking 1994 data having month [1, 8] missing

```python
dfxx[dfxx.index.year == 1994]
```

```python
df[df.index.year == 1994]
```

Looking 2010 data having two 3rd month data

```python
df[df.index.year == 2010]
```

```python
dfxx[dfxx.index.year == 2010]
```

- **Clearly data for 2010-03-31 is doubled, this need to improved**
- **This problem is not encountered anywhere else because only in 2010 months were repeated**

```python
dfxx[(dfxx.index.year == 2010) & (dfxx.index.month == 3)] /= 2
dfxx = dfxx.astype('int')
dfxx[dfxx.index.year == 2010]
```

```python
for year in dfxx.index.year.unique()[1:-1]:
    if len(list(set(dfxx[dfxx.index.year == year].index.month.values))) != 12:
        print(year, sorted(list(set(months) - set(dfxx[dfxx.index.year == year].index.month.values))))
```

```python
dfxx.plot(figsize=(18, 6));
```

**So, all 14 months not having any value is assigned 0 using resample() method**

```python
df.plot(figsize=(18, 6));
```
