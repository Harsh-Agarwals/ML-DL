## Line-charts

A line plot is a type of plot in which data points are represented as a **series of connected line segments**. In Machine Learning and EDA, line plots are often used to **visualize the trends and patterns in a dataset over time or across different values** of an independent variable.


### Uses of Line charts in Machine Learning and EDA:

- **Visualizing time series data**: Line plots are useful for visualizing time series data, where the data points are collected at regular intervals over time. By plotting the data points in order of time, a line plot can **reveal patterns and trends** that might not be apparent from a simple tabular representation of the data. For example, in **stock price analysis**, line plots can be used to visualize the trend of stock prices over time.


- **Exploring the relationship between two variables**: Line plots can also be used to explore the relationship between two variables. For example, in Machine Learning, we may be interested in **visualizing the relationship between a feature and the target variable**. By plotting the feature variable on the x-axis and the target variable on the y-axis, we can see if there is a **linear relationship between the two variables**.


- **Visualizing model performance**: Line plots can be used to *visualize the performance of Machine Learning models over time or across different hyperparameter settings*. By plotting the model **performance metric (such as accuracy or F1 score)** on the y-axis and the hyperparameter values on the x-axis, we can see how the model performs as the hyperparameters are tuned.


- **Identifying outliers**: Outliers are data points that fall far outside the expected range of values for a given variable. *By plotting the data on a line plot, outliers can be easily identified as points that **deviate significantly** from the overall trend of the data*.


Overall, line plots are a powerful tool for visualizing trends and patterns in data, and can be used in a variety of applications in Machine Learning and EDA.

## Line plot with respect to time

```python
### Incorporating the datetime variables
# Using .dt for pandas datetime variables and extracting year, month, week and other datetime variables from those

df.reset_index(inplace=True)
df['Date'] = pd.to_datetime(df['Date'])

df['year'] = df.Date.dt.year
df['month'] = df.Date.dt.month
df['week'] = df.Date.dt.weekofyear
df['day'] = df.Date.dt.dayofyear
df['dom'] = df.Date.dt.day
df['dow'] = df.Date.dt.day_of_week

df
```

## Bar Charts

Bar Charts is a **graphical representation of categorical data** with rectangular bars. Each bar represents a category and the height of the bar represents the value of the corresponding variable.

**Bar plots are commonly used in both Machine Learning and EDA for the following purposes:**

- **Visualizing distribution of categorical variables**: Bar plots are used to visualize the distribution of categorical variables. In Machine Learning, we often have categorical variables in our dataset that we want to analyze. By using bar plots, we can *visualize the distribution of the variables and identify any patterns or trends*.


- **Comparing multiple categories**: Bar plots are useful for comparing multiple categories. In EDA, we may be interested in comparing the performance of different products or services. By plotting the data for each category on a single plot, we can *compare the performance and identify any differences*.


- **Visualizing model performance**: Bar plots can be used to visualize the performance of Machine Learning models. By *plotting the model performance metric (such as accuracy or F1 score)* for each model on a single plot, we can compare the performance of the different models and identify the best performing one.


- **Visualizing frequency distribution**: Bar plots can be used to visualize the frequency distribution of a variable. For example, in EDA, we may be interested in visualizing the frequency distribution of customer age groups or income brackets. By plotting the data in a bar plot, we can *easily visualize the frequency distribution and identify any patterns or anomalies*.


- **Identifying outliers and anomalies**: Bar plots can be used to identify any outliers or anomalies in the data. In EDA, we can use bar plots to visualize the frequency distribution of a variable and **identify any categories that have extremely high or low values**. This can help us identify potential errors in the data or outliers that may need further investigation.

Overall, bar plots are a versatile and useful tool for both Machine Learning and EDA. They can help us visualize the distribution of categorical variables, compare categories, identify outliers and anomalies, and present results in a clear and concise manner.

## Scatter Plots

A scatter plot is a type of plot that is used to **visualize the relationship between two variables**. Each point in a scatter plot represents a *single observation or data point*, and the position of the point on the plot **represents the values of the two variables**. Scatter plots are useful for *identifying patterns or trends* in the data and for *detecting outliers or anomalies*.

**Uses of Scatter Plots:**

- **Visualizing the relationship between variables**: Scatter plots are commonly used in EDA to visualize the relationship between two variables. For example, we might use a scatter plot to *visualize the relationship between a customer's age and their spending habits*.


- **Detecting outliers and anomalies**: Scatter plots can also be used to detect outliers or anomalies in the data. Outliers are observations that are significantly different from the other observations in the dataset, and they can have a significant impact on the results of our analysis.


- **Assessing the linearity of the relationship**: Scatter plots can be used to assess the linearity of the relationship between two variables. *If the points in the scatter plot form a clear linear pattern, it suggests that the two variables have a strong linear relationship*. If the points form a non-linear pattern, it suggests that the relationship may be non-linear.


- **Visualizing clusters and patterns**: In Machine Learning, scatter plots can be used to visualize clusters and patterns in the data. For example, we might use a scatter plot to *visualize the distribution of different classes or labels in the data*, or to *identify patterns that can be used to create new features for our model*.


- **Evaluating model performance**: Scatter plots can be used to evaluate the performance of a machine learning model. For example, we might use a scatter plot to *compare the predicted values of a model to the actual values in the test set*. This can help us identify areas where the model is performing well and areas where it needs improvement.

Overall, scatter plots are a powerful tool for *visualizing the relationship between variables* and *identifying patterns in the data*. By using scatter plots in Machine Learning and EDA, we can **gain insights into our data and develop more accurate and effective models**.

## Heatmaps

A heatmap is a graphical representation of data where the **individual values that are contained in a matrix are represented as colors**. It is a powerful visualization tool used in both Machine Learning and EDA for **understanding the correlation between different variables** in a dataset. Heatmaps provide a *quick and easy way to understand patterns and trends* in the data.

**Uses of Heatmaps:**

- **Visualizing Correlations**: Heatmaps are commonly used to visualize the correlation between variables in a dataset. By visualizing the correlations between variables, we can *quickly identify patterns in the data and understand how different variables are related* to one another. This is useful for **identifying which variables are most important for predicting the target variable**.


- **Identifying Patterns and Trends**: Heatmaps can be used to **identify patterns in the data that might not be easily visible through other visualization methods**. By analyzing the heatmap, we can **identify clusters of similar values that might indicate patterns or trends in the data**. This can be useful for *identifying outliers, trends, and patterns in the data*.


- **Feature Selection**: Heatmaps can also be used for feature selection, which is the process of **selecting the most important features from a dataset**. By analyzing the correlation between different features, we can **identify which features are most important for predicting the target variable**. This can *help in building more accurate and effective models*.


- **Data Comparison**: Heatmaps are useful for comparing data across different categories or time periods. By visualizing the data in a heatmap, we can easily **identify which categories or time periods have the highest or lowest values**. This can be useful for identifying trends and patterns in the data.


- **Identifying Missing Data**: Heatmaps can be used to identify missing data or null values in a dataset. By visualizing the data in a heatmap, we can *easily identify which cells have missing values or null values*. This can be useful for *data cleaning and preprocessing*.

Overall, heatmaps are a powerful tool for visualizing the relationships between variables in a dataset. By using heatmaps in Machine Learning and EDA, we can gain insights into our data and **develop more accurate and effective models**.

```python
### Heatmap for correlation between variables in iris dataset

iris.corr()
```

```python
### Heatmap will simply color code the above values, on the basis of their severity

plt.figure(figsize=(10, 8))
sns.heatmap(iris.corr(), annot=True)
plt.show()
```

**Observations:**
- Correlation of 1 with itself
- sepal_length shows negative correlation with sepal width (also we saw that using regression plot)
- Sepal length shows high positive correlation with petal length and petal width
- Sepal width is negatively correlated with all the other features
- Petal length has high correlation with petal width, same was not in case of sepal length and width
